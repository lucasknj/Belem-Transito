#### Define
As colunas que dizem respeito à percepção dos moradores de forma quantitativa (notas de 1 a 5) estão sendo representadas ao contrário. Notas 1 estão sendo consideradas boas e notas 5 consideradas péssimas.
Se formos, por exemplo, medir a escala de satisfação de um produto, e usar-mos a classificação de 1 a 5, produtos com média alta indicariam maior satisfação.

As novas colunas para guardar essas informações serão nota_comportamento e nota_container

```python
d['nota_comportamentoDosMoradores'] = 0.0
```

```python
for x in d['Sobre o comportamento dos moradores acerca da dispensa do lixo em locais adequados, NA SUA RUA, qual sua impressão:']:
    print (x)
    if x==5.0:
        d.loc[d['Sobre o comportamento dos moradores acerca da dispensa do lixo em locais adequados, NA SUA RUA, qual sua impressão:']==x,'nota_comportamentoDosMoradores']=1.0
    elif x==4.0:
        d.loc[d['Sobre o comportamento dos moradores acerca da dispensa do lixo em locais adequados, NA SUA RUA, qual sua impressão:']==x,'nota_comportamentoDosMoradores']=2.0
    elif x==3.0:
        d.loc[d['Sobre o comportamento dos moradores acerca da dispensa do lixo em locais adequados, NA SUA RUA, qual sua impressão:']==x,'nota_comportamentoDosMoradores']=3.0
    elif x==2.0:
        d.loc[d['Sobre o comportamento dos moradores acerca da dispensa do lixo em locais adequados, NA SUA RUA, qual sua impressão:']==x,'nota_comportamentoDosMoradores']=4.0
    elif x==1.0:
        d.loc[d['Sobre o comportamento dos moradores acerca da dispensa do lixo em locais adequados, NA SUA RUA, qual sua impressão:']==x,'nota_comportamentoDosMoradores']=5.0
```

```python
d['nota_comportamentoDosMoradores'].value_counts()
```

```python
d['nota_depositoDeLixo'] = 0.0
```

```python
for x in d['Sobre os locais e contêineres para deposito de lixo oferecidos pela Prefeitura para descarte, avalie:']:
    print (x)
    if x==5.0:
        d.loc[d['Sobre os locais e contêineres para deposito de lixo oferecidos pela Prefeitura para descarte, avalie:']==x,'nota_depositoDeLixo']=1.0
    elif x==4.0:
        d.loc[d['Sobre os locais e contêineres para deposito de lixo oferecidos pela Prefeitura para descarte, avalie:']==x,'nota_depositoDeLixo']=2.0
    elif x==3.0:
        d.loc[d['Sobre os locais e contêineres para deposito de lixo oferecidos pela Prefeitura para descarte, avalie:']==x,'nota_depositoDeLixo']=3.0
    elif x==2.0:
        d.loc[d['Sobre os locais e contêineres para deposito de lixo oferecidos pela Prefeitura para descarte, avalie:']==x,'nota_depositoDeLixo']=4.0
    elif x==1.0:
        d.loc[d['Sobre os locais e contêineres para deposito de lixo oferecidos pela Prefeitura para descarte, avalie:']==x,'nota_depositoDeLixo']=5.0
```
